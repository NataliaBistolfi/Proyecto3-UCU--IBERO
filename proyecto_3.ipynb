## Gráficas de comparación PBI vs Tasa de Desempleo

## Gráfica de comparación de México
#### Ingreso de datos de comparación presentado en un gráfico de doble eje a diferente escala al eje y (vertical), y la misma escala en el eje x (horizontal).

```python
# Datos del PBI de México
years_pbi = df_pbi_mexico.index
pbi_values = df_pbi_mexico['Mexico']

# Datos de la tasa de desempleo de México
years_unemployment = df_mexico.index
unemployment_values = df_mexico['México']

# Creamos la figura y los ejes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Configuramos el eje izquierdo para el PBI
ax1.set_xlabel('Años')
ax1.set_ylabel('PBI (%)', color='blue')
ax1.plot(years_pbi, pbi_values, label='PBI México', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creamos el segundo eje para la tasa de desempleo
ax2 = ax1.twinx()  
ax2.set_ylabel('Tasa de desempleo (%)', color='green')
ax2.plot(years_unemployment, unemployment_values, label='Desempleo México', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Título y leyendas
fig.suptitle('PBI vs Tasa de Desempleo en México (2005-2023)')
fig.tight_layout()  
plt.show()
```

## Conclusión final

```python
#El análisis muestra que un crecimiento del PBI genera una disminución significativa en la tasa de desempleo, pero no siempre de forma inmediata. Este desfase temporal se debe a que la relación entre estas dos variables no es directa, sino que depende de diversos factores económicos. Por ejemplo, cuando el PBI aumenta, las empresas tienden a incrementar su producción, lo que eventualmente impulsa la contratación de más trabajadores. Sin embargo, este proceso puede tardar debido a las dinámicas de inversión, ajustes en la capacidad productiva y la respuesta del mercado laboral.

#En general, se observa que, tras un periodo de crecimiento sostenido del PBI, la tasa de desempleo tiende a disminuir de manera considerable, validando la conexión entre la expansión económica y la mejora del mercado laboral. Este comportamiento refuerza la importancia de políticas que promuevan el crecimiento económico como estrategia para reducir el desempleo.
```
